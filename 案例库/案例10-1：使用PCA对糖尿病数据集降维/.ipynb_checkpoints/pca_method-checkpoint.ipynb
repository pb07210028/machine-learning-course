# 1. 数据预处理

## 1.1 导入包

```python
# -*- coding: utf-8 -*-
"""
Python 3.7.7
sklearn 0.23.1
使用 PCA 对糖尿病数据集降维
"""

# 导入包
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## 1.2 导入数据集

```python
# 导入数据集
dataset = pd.read_csv('pima-indians-diabetes.csv')
print(dataset.shape)
```

## 1.3 检测缺失值

```python
# 检测缺失值
null_df = dataset.isnull().sum()
print(null_df)
```

## 1.4 生成自变量和因变量

```python
# 生成自变量和因变量
X = dataset.iloc[:,0:8].values
y = dataset.iloc[:,8].values
```

## 1.5 拆分训练集和测试集

```python

# 拆分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
print(X_train.shape)
```

## 1.6 特征缩放

```python
# 特征缩放
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# 2. 使用PCA生成新的自变量

## 2.1 使用 PCA 生成新的自变量

```python
# 使用 PCA 生成新的自变量
pca = PCA(n_components = None) # 不指定新的自变量的个数
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
print(X_train_pca)
```

## 2.2 打印旧的自变量与新的自变量的转换系数

```python
# 打印旧的自变量与新的自变量的转换系数
print('打印旧的自变量与新的自变量的转换系数是：\n', pca.components_)
```

## 2.3 增加转换系数的可读性

```python
# 增加转换系数的可读性
old_columns = list(dataset)[:-1]
new_columns = ['pc' + str(i) + '_component' for i in range(X_train.shape[1])]
components_df = pd.DataFrame(pca.components_, 
                             columns = old_columns, 
                             index = new_columns)
components_df = components_df.T # 转置，增加可读性
print(components_df)
```

## 2.4 检验X_train_pca的由来

```python
# 检验x_train_pca的由来
components = components_df.values
verify_matrix = X_train.dot(components) # verify_matrix 和 X_train_pca一模一样
```

```python
print(verify_matrix)
```

```python
print(X_train_pca)
```

#### verify_matrix 和 X_train_pca一模一样

## 2.5 打印 pca 的方差解释比率

```python
# 打印 pca 的方差解释比率
print('PCA的方差解释比率是：\n', pca.explained_variance_ratio_)
```

## 2.6 画出新的自变量的个数 VS 累计方差解释

```python
# 画出新的自变量的个数 VS 累计方差解释
plt.plot([i for i in range(1, X_train.shape[1] + 1)], np.cumsum(pca.explained_variance_ratio_), c='orange')
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

#### 当降维后的自变量的个数是6时，能解释90%的方差。所以选择降维后自变量的个数是6。

## 2.7 使用 PCA 降维

```python
# 使用 PCA 降维
pca = PCA(n_components = 6) # 6由上一步选出
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
print(X_train_pca)
```

# 3. 构建逻辑回归模型

## 3.1 使用原始数据构建逻辑回归模型

```python
# 构建模型
classifier = LogisticRegression(penalty='l2', C=1, class_weight='balanced', random_state = 0)
classifier.fit(X_train, y_train)

# 预测测试集
y_pred = classifier.predict(X_test)

# 评估模型性能
print(accuracy_score(y_test, y_pred))
```

## 3.2 使用降维后数据构建逻辑回归模型

```python
# 构建模型
classifier = LogisticRegression(penalty='l2', C=1, class_weight='balanced', random_state = 0)
classifier.fit(X_train_pca, y_train)

# 预测测试集
y_pred = classifier.predict(X_test_pca)

# 评估模型性能
print(accuracy_score(y_test, y_pred))
```

#### 降维后，模型性能提升了0.6%

# 4. 可视化PCA降维效果

#### 可视化时，选择2个主成分。选择2个主成分信息有损失，这里目的仅仅是可视化

## 4.1 选择2个主成分

```python
# 使用 PCA 降维
pca = PCA(n_components = 2)
X_train_pca = pca.fit_transform(X_train)
```

## 4.2 可视化

```python
X_set, y_set = X_train_pca, y_train
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('PCA Viz')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

经过PCA降维，自变量由8个变为2个。
将降维后的2个主成分可视化，可以看到，如果以2个主成分训练逻辑回归模型，模型性能会较差，因为肉眼可见，2个类别之间没有明显的界限。
