# 1. 导入包

```python
'''
Python 3.7.7
sklearn 0.23.1
使用K-Means对超市客户分组
'''

# 导入包
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

```

# 2. 导入数据集

#### 为了可视化聚类效果，仅选取Annual Income (k$)和Spending Score (1-100)这2个字段

```python
# 导入数据集
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

# 3. 使用肘部法则选择最优的K值

```python
# 使用肘部法则选择最优的K值
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', n_init=10, max_iter=300, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

# 4. 画出 聚类个数 vs WCSS 图

```python
# 画出 聚类个数 vs WCSS 图
plt.figure()
plt.plot(range(1, 11), wcss, 'ro-')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### 从K=5开始，WCSS下降的不再明显，说明K=5是最优选择

# 5. 使用K=5做聚类

```python
# 使用选择出的K，使用K-Means做聚类
kmeans = KMeans(n_clusters = 5, init = 'k-means++', n_init=10, max_iter=300, random_state = 0)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

# 6. 可视化聚类效果

```python
# 可视化聚类效果
plt.figure()
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

# 7. 采取措施

#### Cluster 1 工资收入中等，消费中等；
#### Cluster 2 工资收入低，消费高，查看这个分组主要购买哪些商品；
#### Cluster 3 工资收入高，消费高；
#### Cluster 4 工资收入低，消费低；
#### Cluster 5 工资收入高，消费低，给这个分组的客户办理优惠券或打折购物卡，吸引他们消费；
