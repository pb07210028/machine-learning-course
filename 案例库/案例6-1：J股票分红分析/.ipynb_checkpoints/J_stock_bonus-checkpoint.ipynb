# 1. 导入包

```python
'''
基于时间序列预测J股票未来两年分红
'''

# 导入包
import pandas as pd
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf 
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.gofplots import qqplot
import matplotlib.pyplot as plt
from statsmodels.tsa import arima_model
from statsmodels.stats.diagnostic import acorr_ljungbox
```

# 2. 导入数据集

```python
# 导入数据集
mydata = pd.read_csv('J_stock_bonus.csv')
```

# 3. 创建时序数据

```python
# 创建时序数据
ts=pd.period_range('1997/01',periods=len(mydata),freq='Q')
bonus_data=pd.DataFrame(mydata['bonus'].values,index=ts,columns=['bonus'])
bonus_data
```

```python
# 绘制时间序列图
plt.figure()
bonus_data.plot()
plt.xlabel('date')
plt.ylabel('bonus')
plt.show()
```

#### 数据不符合加法模型，需要做转换

```python
# 时间序列转换
bonus_log=np.log(bonus_data)
plt.figure()
bonus_log.plot()
plt.xlabel('date')
plt.ylabel('log(bonus)')
plt.show()
```

#### 转换后的数据适用于加法模型

# 4. 平稳性检验

```python
# 平稳性检验
adfuller(bonus_log['bonus'])
```

#### p值=0.8179273139117222，大于0.05，说明不平稳，需要通过差分使其变得平稳

# 5. 差分序列

```python
# 差分序列
bonus_log_diff1 = bonus_log.diff().dropna()
# 绘制差分后的时间序列图
plt.figure()
bonus_log_diff1.plot()
plt.xlabel('date')
plt.ylabel('log(bonus) diff1')
plt.show()
```

```python
# 差分后再次做平稳性检验
adfuller(bonus_log_diff1['bonus'])
```

#### p值=0.0004149732205762509，小于0.05，代表数据已经平稳

# 6. 白噪声检验

```python
# 白噪声检验
df = acorr_ljungbox(bonus_log_diff1['bonus'], lags = [6, 12], boxpierce=True, return_df = True)
df
```

#### array([1.19918483e-18, 7.64609294e-29])，p值很小，该数据无法拒绝原假设，即认为该数据不是随机数据，即该数据有可分析价值

# 7. 查看自相关图和偏自相关图

```python
# 查看自相关图
plt.figure()
plot_acf(bonus_log_diff1,lags=30) # lags必须小于一半数据
plt.xlabel('LAG')
plt.ylabel('ACF')
plt.show()
```

#### ACF重现拖尾现象，并且在第1阶或低4阶呈现衰减，因此取q=1或4

```python
# 查看偏自相关图
plt.figure()
plot_pacf(bonus_log_diff1,lags=30)
plt.xlabel('LAG')
plt.ylabel('PACF')
plt.show()
```

#### PACF重现拖尾现象，并且在第3阶或低4阶呈现衰减，因此取p=3或4

#### (p,d,q)取值有(3,1,1), (3,1,4), (4,1,1), (4,1,4)

#### 由于ACF是拖尾，PACF也是拖尾，所以使用ARMA/ARIMA模型，而不能使用AR/MA模型

# 8. 找到最优p和q

```python
# 找到最优p, q
# 生成一个空列表，用于存储AIC值
aic_values=[]
# 初始化变量
find_index=0
for p in range(8):
    for q in range(8):
        try:
            myfit=arima_model.ARIMA(bonus_log,(p,1,q)).fit()
            aic_values.append(myfit.aic)
            print('(%d:%d,1,%d),AIC:%d'%(find_index,p,q,myfit.aic))
            find_index+=1
        except:
            pass
# 打印最小的AIC值
print(aic_values.index(min(aic_values)),min(aic_values))
```

#### 找AIC最小的记录。发现 (23:3,1,2),AIC:-151 是期望的记录。23代表序号，3代表p，1代表d，2代表q。

# 9. 构建模型

```python
# 建立模型,选择模型为arima(3,1,2)
arima_fit=arima_model.ARIMA(bonus_log, order=(3,1,2)).fit()
```

```python
arima_fit.summary()   
```

# 10. 残差检验，观察是否符合正态分布

```python
# 残差检验，观察是否符合正态分布
# 绘制残差自相关图
plot_acf(arima_fit.resid)
adfuller(arima_fit.resid)
```

```python
# 绘制QQ图
qqplot(arima_fit.resid,line='s')
plt.figure()
```

```python
# 绘制正态分布图
plt.hist(arima_fit.resid,density=True)
arima_fit.resid.plot(kind='kde')
plt.show()
```

#### 主要通过QQ图查看是否符合正态分布。对于QQ图，线性即正态分布，所以数据是符合正态分布的，说明残差是随机数据

# 11. 预测未来两年的股票分红

```python
# 预测后8个季度（未来两年）的股票分红。（由于ARIMA模型需要的是经过对数化的数据，所以需要通过指数化进行数据还原）
bonus_forecast=np.exp(arima_fit.forecast(8)[0])
print(bonus_forecast)
```

# 12. 可视化原始数据与预测结果

```python
# 可视化原始数据与预测结果
plt.figure()
plt.plot(range(len(bonus_data)),bonus_data['bonus'],label='bonus')
plt.plot(range(len(bonus_data),len(bonus_data)+len(bonus_forecast)),
         bonus_forecast,label='forecast',color='r')
plt.xlim(0,100)
plt.ylim(0,25)
plt.legend()
plt.show()
```

#### 红色折线图是预测后的结果。可以看到大体是符合趋势的。
