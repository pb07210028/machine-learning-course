# 1. 数据预处理

## 1.1 导入包

```python
# -*- coding: utf-8 -*-
"""
Python 3.7.7
sklearn 0.23.1
使用 LDA 对糖尿病数据集降维
"""

# 导入包
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

## 1.2 导入数据集

```python
# 导入数据集
dataset = pd.read_csv('pima-indians-diabetes.csv')
print(dataset.shape)
```

## 1.3 检测缺失值

```python
# 检测缺失值
null_df = dataset.isnull().sum()
print(null_df)
```

## 1.4 生成自变量和因变量

```python
# 生成自变量和因变量
X = dataset.iloc[:,0:8].values
y = dataset.iloc[:,8].values
```

## 1.5 拆分训练集和测试集

```python

# 拆分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
print(X_train.shape)
```

## 1.6 特征缩放

```python
# 特征缩放
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# 2. 使用 LDA 生成新的自变量

```python
# 使用 LDA 生成新的自变量
lda = LDA(n_components = None)
X_train_lda = lda.fit_transform(X_train, y_train)
X_test_lda = lda.transform(X_test)
print(X_train_lda)
```

如果 n_components 的值是None，那么其值被设置为 min(n_classes - 1, n_features)，即min(1,8)=1

```python
lda.scalings_
```

# 3. 构建逻辑回归模型

## 3.1 使用原始数据构建逻辑回归模型

```python
# 构建模型
classifier = LogisticRegression(penalty='l2', C=1, class_weight='balanced', random_state = 0)
classifier.fit(X_train, y_train)

# 预测测试集
y_pred = classifier.predict(X_test)

# 评估模型性能
print(accuracy_score(y_test, y_pred))
```

## 3.2 使用降维后数据构建逻辑回归模型

```python
# 构建模型
classifier = LogisticRegression(penalty='l2', C=1, class_weight='balanced', random_state = 0)
classifier.fit(X_train_lda, y_train)

# 预测测试集
y_pred = classifier.predict(X_test_lda)

# 评估模型性能
print(accuracy_score(y_test, y_pred))
```

降维后，模型性能和降维之前一样。

# 4. 可视化LDA降维效果

由于X_train_lda是1维的。为了可视化，为X_train_lda扩展1维

```python
to_insert = np.ones((614,1))
X_train_lda = np.c_[X_train_lda, to_insert]
```

```python
X_set, y_set = X_train_lda, y_train
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('PCA Viz')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

经过LDA降维，自变量由8个变为1个。 肉眼可见，模型性能不错，因为红色的类别集中在左边，绿色的类别集中在右边。
